```python
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([45, 50, 55, 60, 65])

# Lines with different intercepts
lines = {
    "Intercept 40": 5 * X + 40,
    "Intercept 44": 5 * X + 44,
    "Intercept 35": 5 * X + 35
}

# Plot actual data points
plt.scatter(X, y, color='blue', label='Actual Data', s=80)

# Plot lines
colors = ['red', 'green', 'orange']
for (label, y_line), color in zip(lines.items(), colors):
    plt.plot(X, y_line, label=label, color=color, linewidth=2)

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Effect of Changing Intercept on Fit')
plt.legend()
plt.grid(True)
plt.show()
```
