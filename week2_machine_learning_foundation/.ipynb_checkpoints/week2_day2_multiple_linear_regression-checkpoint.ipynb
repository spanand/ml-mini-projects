```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Data
data = {
    'Hours_Studied': [1, 2, 3, 4, 5],
    'Hours_Slept': [6, 7, 6, 8, 7],
    'Exam_Score': [50, 60, 65, 80, 85]
}

df = pd.DataFrame(data)
X = df[['Hours_Studied', 'Hours_Slept']]
y = df['Exam_Score']

# Model
model = LinearRegression()
model.fit(X, y)

# Coefficients
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Prediction example
pred = model.predict([[3, 7]])  # 3 hrs studied, 7 hrs slept
print("Predicted Exam Score:", pred[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Data
X = np.array([
    [0.5, 6.35],
    [1, 6],
    [2, 7],
    [3, 6],
    [4, 8],
    [5, 7]
])
y = np.array([46.44,50, 60, 65, 80, 85])

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict
# Prediction example
pred = model.predict([[0.5, 6.35]])  # 3 hrs studied, 7 hrs slept
print("Predicted Exam Score:", pred[0])

pred = model.predict([[3, 7]])
print("Predicted Score:", pred[0])

# Create meshgrid for 3D surface
x_surf, y_surf = np.meshgrid(
    np.linspace(X[:, 0].min(), X[:, 0].max(), 10),
    np.linspace(X[:, 1].min(), X[:, 1].max(), 10)
)
z_surf = model.predict(
    np.column_stack((x_surf.ravel(), y_surf.ravel()))
).reshape(x_surf.shape)

# Plot 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")

# Actual data points
ax.scatter(X[:, 0], X[:, 1], y, color='blue', s=60, label='Actual Data')

# Regression surface
ax.plot_surface(x_surf, y_surf, z_surf, color='orange', alpha=0.6)

ax.set_xlabel('Hours Studied')
ax.set_ylabel('Hours Slept')
ax.set_zlabel('Exam Score')
ax.set_title('3D Regression Plane - Multiple Linear Regression')
ax.legend()

plt.show()
```

```python
#Model Evaluation & Metrics
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Predictions on training data
y_pred = model.predict(X)

# Metrics
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)  # compute RMSE manually
r2 = r2_score(y, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R²:", r2)
```

```python
#Residuals
```

```python
import matplotlib.pyplot as plt

# Residuals
residuals = y - y_pred

# Plot
plt.figure(figsize=(8,5))
plt.scatter(range(len(y)), residuals, color='red', s=60)
plt.hlines(y=0, xmin=0, xmax=len(y)-1, colors='blue', linestyles='dashed')
plt.xlabel('Data Point Index')
plt.ylabel('Residual (Actual - Predicted)')
plt.title('Residual Plot - Multiple Linear Regression')
plt.grid(True)
plt.show()
```
